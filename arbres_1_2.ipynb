```python
anne_genre_sum = {}
for _, i, count in anne_genre_result_sorted:
    if i in anne_genre_sum:
        anne_genre_sum[i] += count
    else:
        anne_genre_sum[i] = count
print(anne_genre_sum)
anne_genre_sum_list = list(anne_genre_sum.values())
print(anne_genre_sum_list)
```

```python
s =[]
r = []
for t in anne_genre_result_sorted:
    for key in anne_genre_sum.keys():
        if key in range(1950, t[1]+1):
            s.append(anne_genre_sum[key])
            totale = sum(s)
            print(totale)
            if totale != 0:
                percentage = t[2] / totale * 100
                r.append((t[0],t[1], percentage))
            else:
                r.append((t[0],t[1], 0))
```

```python
import matplotlib.pyplot as plt
years = [item[1] for item in r]
percentages = [item[2] for item in r]
labels = [item[0] for item in r]
plt.figure(figsize=(10, 6))
plt.scatter(years, percentages, color='blue')

for i, label in enumerate(labels):
    plt.text(years[i], percentages[i], label, fontsize=1, ha='right')

plt.title('Percentage by Year for Each Species')
plt.xlabel('Year')
plt.ylabel('Percentage')

plt.grid(True)
plt.show()

```
