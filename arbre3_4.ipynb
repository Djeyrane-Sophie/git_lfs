# Quelle est la proportion de chacun des genres botaniques par rapport au total ?

```python
somme_genre_panda = count_genre['count_g'].sum()
count_genre['pourcentage'] = count_genre.apply(lambda row: row['count_g']/ somme_genre_panda *100 , axis=1)
print(count_genre)
```

# Les objectifs de la Métro sont-ils respectés en termes de diversité ?

```python
genre_espece_pandas = df['GENRE_BOTA']+" " + df['ESPECE']
genre_espece_count_pandas  = genre_espece_pandas .value_counts().reset_index()
genre_espece_count_pandas.columns = ['GENRE_ESP','count_e_g']
somme_genre_espece_panda = count_genre['count_g'].sum()
genre_espece_count_pandas['pourcentage'] = genre_espece_count_pandas.apply(lambda row: row['count_e_g']/ somme_genre_espece_panda *100 , axis=1)
print(genre_espece_count_pandas)
```

```python
cg = count_genre.head(1)
gecp = genre_espece_count_pandas.head(1)
```

```python
gecp['obj_metro'] = gecp['pourcentage'].apply(lambda row: 'Oui' if row< 10 else 'Non' )
cg['obj_metro'] = cg['pourcentage'].apply(lambda row: 'Oui' if row< 20 else 'Non' )
```
